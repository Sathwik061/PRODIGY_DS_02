```python
# Install libraries if not already done
# pip install pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Titanic training data
titanic_df = pd.read_csv("C:\\Users\\madire sathwik\\Downloads\\titanic\\train.csv")
```

```python
# Fill missing 'Age' values with median
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())



# Fill missing 'Embarked' with the most frequent value
titanic_df['Embarked'] = titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0])

# Drop 'Cabin' column only if it exists
if 'Cabin' in titanic_df.columns:
    titanic_df.drop(columns=['Cabin'], inplace=True)
```

```python
# Set Seaborn style
sns.set(style="whitegrid")
```

```python
# --- Survival Count ---
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=titanic_df)
plt.title('Survival Count')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.ylabel("Number of Passengers")
plt.show()
```

```python
# --- Survival by Gender ---
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', hue='Survived', data=titanic_df)
plt.title('Survival by Gender')
plt.ylabel("Count")
plt.legend(labels=['Not Survived', 'Survived'])
plt.show()
```

```python
# --- Survival by Passenger Class ---
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', hue='Survived', data=titanic_df)
plt.title('Survival by Passenger Class')
plt.ylabel("Count")
plt.legend(labels=['Not Survived', 'Survived'])
plt.show()
```

```python
# --- Age Distribution Histogram ---
plt.figure(figsize=(8, 4))
sns.histplot(data=titanic_df, x='Age', bins=30, kde=True, hue='Survived', palette='Set2')
plt.title('Age Distribution by Survival')
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
# --- Gender Distribution Bar Chart ---
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', data=titanic_df)
plt.title('Gender Distribution of Passengers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# --- Correlation Heatmap ---
plt.figure(figsize=(10, 6))
sns.heatmap(titanic_df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# --- Survival by Embarked Port ---
plt.figure(figsize=(6, 4))
sns.countplot(x='Embarked', hue='Survived', data=titanic_df)
plt.title('Survival by Port of Embarkation')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.legend(labels=['Not Survived', 'Survived'])
plt.show()
```
